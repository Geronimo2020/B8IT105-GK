```python
# Just chasing a lead regarding the data and how it is split - checking whether  the sum of Total Volume 
# equals is the same as the totals of all the individual years summed. From my quick look at this there seems to 
# be some discrepancy


data2015 = df[df.year==2015]
data2016 = df[df.year==2016]
data2017 = df[df.year==2017]
data2018 = df[df.year==2018]
sum2015 = data2015['TotalVolume'].sum()
sum2016 = data2016['TotalVolume'].sum()
sum2017 = data2017['TotalVolume'].sum()
sum2018 = data2018['TotalVolume'].sum()
sumTotal = sum2015 + sum2016 + sum2017 + sum2018 
sumTotalVolume = df.TotalVolume.sum()


print('The total sales volume in 2015: ', sum2015)
print('The total sales volume in 2016: ',sum2016)
print('The total dales volume in 2017: ',sum2017)
print('The total dales volume in 2018: ',sum2018, '  * note this was a truncated year in terms of data collected')
print('The total dales volume BY ADDDING YEAR DATA: ',sumTotal)
print('The total dales volume BY SUMMING THE TOTAL VOLUME COLUMN: ', sumTotalVolume)

print('The difference between the year summation and the TotalVolum sum is : ',sumTotal - sumTotalVolume)
```

```python
# Quick investigation that the region TotalUs is not mistaenly just the aggregated total 
# of the dataset as a whole and added back into data


## ResulVolume without the Total US summed figure
## The result  proves that TotalUS region is not simply the total for the whole US added into 
## the data as a region variable. 

dataTotalUS = df[df.region=='TotalUS'] # 
sumTotalVolume = df['TotalVolume'].sum()# Total for whole data set
sumTotalUS = dataTotalUS ['TotalVolume'].sum() # Total for the USTotal region
Total_minusTotalUS = sumTotalVolume  - sumTotalUS # Dataset total minus TotalUS total

print('The summed Total Volume  for the region TotalUS = ', sumTotalUS)

print('The summed Total Volume minus the summed total for the region TotalUS = ', Total_minusTotalUS) 

```
